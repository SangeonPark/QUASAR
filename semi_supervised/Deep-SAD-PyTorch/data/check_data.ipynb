```python
data = np.array(pd.read_hdf("Nsubjettiness_RandD.h5"))
```

```python
for i in range(1,13):
    data[:,i] = (data[:,i]-np.mean(data[:,i]))/np.std(data[:,i])
```

```python
data = np.column_stack((data[:,0:1],data[:,1:13],data[:,-1:]))
```

```python
data_pd = pd.DataFrame(data=data)
```

```python
data_pd.head()
```

```python
data_pd.to_hdf("Nsubjettiness_RandD_normalized.h5",key="normalized")
```
